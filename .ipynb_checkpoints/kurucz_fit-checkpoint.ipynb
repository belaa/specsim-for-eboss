#### Unit normalization fit

```python
# Select fluxes that lie between 3000-10000 angstroms
def keep_visible(kurucz):
    
    '''
    Returns wavelength in angstroms, and flux in erg/cm^2/sec/angstrom
    '''
    
    keep = np.logical_and(kurucz['col1']*10 > 3000, kurucz['col1']*10 < 12000)
    return kurucz['col1'][keep]*10, kurucz['col2'][keep]


# Normalize spectra to 1
def normalize(flux):
    return flux/np.sum(flux)

# Set up grid of possible alpha values
alpha = np.arange(0,0.2,0.001)
```

```python
def float_normalization(model_flux, data_flux, ivar):
    alpha_num = np.sum(model_flux*data_flux*ivar)
    alpha_den = np.sum((model_flux**2)*ivar)
    return(alpha_num/alpha_den)

def compute_chisq(data, alpha, model, ivar):
    num = (data-alpha*model)**2
    return np.sum(num*ivar)
```

### Input Kurucz, get fit

```python
def fit_kurucz(stellar_model, data):
    
    '''Stellar model is a list, data is a single array.'''
    
    e_wlen, e_flux, e_ivar = 10**data['loglam'], data['flux'], data['ivar']
    min_chisq = 1000000
    best_alpha = 0
    best_model = np.empty(shape(e_wlen), dtype='float')
    
    for i, model in enumerate(stellar_model):
    
        t = Table.read(os.path.join(path, model), format='ascii')
        # first interpolate model over data wavelength grid 
        k_wlen, k_flux = keep_visible(t)
        model_flux = np.interp(x=e_wlen, xp=k_wlen, fp=k_flux)
        
        # normalize 
        alpha = float_normalization(model_flux, e_flux, e_ivar)
        chi_sq = compute_chisq(e_flux, alpha, model_flux, e_ivar)
        if chi_sq < min_chisq:
            min_chisq = chi_sq
            best_alpha = alpha
            best_model = model_flux
            
    
    return(min_chisq, alpha, e_wlen, e_flux, best_model)
```

```python
def new_fit_kurucz(stellar_model, data):
    
    # data
    e_wlen, e_flux, e_ivar = 10**data['loglam'], data['flux'], data['ivar']
    
    # interpolate model over data wavelength grid
    k_wlen, k_flux = keep_visible(stellar_model)
    model_flux = np.interp(x=e_wlen, xp=k_wlen, fp=k_flux)
    
    #normalize
    alpha = float_normalization(model_flux, e_flux, e_ivar)
    chi_sq = compute_chisq(e_flux, alpha, model_flux, e_ivar)
    
    return(chi_sq, alpha, e_wlen, e_flux, model_flux)
```

```python
min_chisq, alpha, e_wlen, e_flux, model_flux = fit_kurucz([km01_7000_g40_7120, km10_5250_g00_5270, km20_5750_g20_5870], 
                                                  eboss_spec)
```

```python
plt.scatter(e_wlen, e_flux, s=1, label='data')
plt.scatter(e_wlen, alpha*model_flux, s=1, label='model (alpha={:.2f})'.format(alpha))
plt.legend()
plt.show()
```

## Start here

```python
# Select fluxes that lie between 3000-10000 angstroms
def keep_visible(kurucz):
    
    '''
    Returns wavelength in angstroms, and flux in erg/cm^2/sec/angstrom
    '''
    
    keep = np.logical_and(kurucz['col1']*10 > 3000, kurucz['col1']*10 < 12000)
    return kurucz['col1'][keep]*10, kurucz['col2'][keep]


def float_normalization(model_flux, data_flux, ivar):
    alpha_num = np.sum(model_flux*data_flux*ivar)
    alpha_den = np.sum((model_flux**2)*ivar)
    return(alpha_num/alpha_den)

def compute_chisq(data, alpha, model, ivar):
    num = (data-alpha*model)**2
    return np.sum(num*ivar)


def fit_kurucz(kurucz_table, data):
    
    # data
    e_wlen, e_flux, e_ivar = 10**data['loglam'], data['flux'], data['ivar']
    
    # interpolate model over data wavelength grid
    k_wlen, k_flux = keep_visible(kurucz_table)
    model_flux = np.interp(x=e_wlen, xp=k_wlen, fp=k_flux)
    
    #normalize
    alpha = float_normalization(model_flux, e_flux, e_ivar)
    chi_sq = compute_chisq(e_flux, alpha, model_flux, e_ivar)
    
    return(chi_sq, alpha, e_wlen, e_flux, model_flux)


def find_best_fit():

    path = '/Users/belaabolfathi/Desktop/kurucz/'
    eboss_spec = Table.read('/Users/belaabolfathi/Downloads/spec-6755-56413-0024.fits')

    min_chisq = 1000000
    best_alpha = 0
    best_model = np.empty(shape(eboss_spec['flux']), dtype='float')


    for kurucz in os.listdir(path):
        if kurucz.startswith('k'):
            kurucz_table = Table.read(os.path.join(path, kurucz), format='ascii')
            chi_sq, alpha, e_wlen, e_flux, model_flux = fit_kurucz(kurucz_table, eboss_spec)
            if chi_sq < min_chisq:
                min_chisq = chi_sq
                best_alpha = alpha
                best_model = model_flux
    return(min_chisq, best_alpha, best_model, e_wlen, e_flux)
```

```python
min_chisq, best_alpha, best_model, e_wlen, e_flux = find_best_fit()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(e_wlen, e_flux, s=1, alpha=0.3)
plt.scatter(e_wlen, best_alpha*best_model, s=1)
plt.show()
```
