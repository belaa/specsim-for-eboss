### Access spCFrame files to get:

* Wavelength
* Wavelength dispersion
* Row width

```python
def get_cframe_hdulist(plate, mjd, band, fiber, sequence_number=0): #sequence_number == expidx?
    
    '''Prepare plan path to access HDUList for a specific camera (b1, b2, r1, r2).'''

    plan_path = finder.get_plate_plan_path(plate=plate, mjd=mjd, combined=True)
    local_plan_path = local_plan_path = mirror.get(plan_path)
    Plan = bossdata.plate.Plan(local_plan_path)
    exp_name = Plan.get_exposure_name(sequence_number=sequence_number, band=band, fiber=fiber, ftype='spCFrame')
    local_fits = mirror.get(os.path.join(plan_path.strip(os.path.basename(plan_path)), exp_name))
    return(fits.open(local_fits))

def get_cframe_data(plate, mjd, sequence_number=1):
    
    '''Access cFrame data for each camera.'''
    
    b1_hdulist = get_cframe_hdulist(plate=plate, mjd=mjd, band='blue', fiber=30, sequence_number=sequence_number)
    b2_hdulist = get_cframe_hdulist(plate=plate, mjd=mjd, band='blue', fiber=600, sequence_number=sequence_number)
    r1_hdulist = get_cframe_hdulist(plate=plate, mjd=mjd, band='red', fiber=30, sequence_number=sequence_number)
    r2_hdulist = get_cframe_hdulist(plate=plate, mjd=mjd, band='red', fiber=600, sequence_number=sequence_number)
    
    return b1_hdulist, b2_hdulist, r1_hdulist, r2_hdulist

def get_neff(plate, mjd):
    
    '''Get profile width data for all fibers for each camera.'''
    
    remote_plate_spec_path = finder.get_plate_spec_path(plate=plate, mjd=mjd)
    plate_spec_path = mirror.get(remote_plate_spec_path)
    PlateFile = bossdata.plate.PlateFile(plate_spec_path)    
    
    exp_nameb1 = PlateFile.exposures.get_exposure_name(exposure_index=1, camera='b1', 
                                                     ftype='spFlat')
    exp_nameb2 = PlateFile.exposures.get_exposure_name(exposure_index=1, camera='b2', 
                                                     ftype='spFlat')
    exp_namer1 = PlateFile.exposures.get_exposure_name(exposure_index=1, camera='r1', 
                                                     ftype='spFlat')
    exp_namer2 = PlateFile.exposures.get_exposure_name(exposure_index=1, camera='r2', 
                                                     ftype='spFlat')

    remote_plate_path_b1 = finder.get_plate_path(plate=plate, filename=exp_nameb1)
    remote_plate_path_b2 = finder.get_plate_path(plate=plate, filename=exp_nameb2)
    remote_plate_path_r1 = finder.get_plate_path(plate=plate, filename=exp_namer1)
    remote_plate_path_r2 = finder.get_plate_path(plate=plate, filename=exp_namer2)
    
    plate_path_b1 = mirror.get(remote_plate_path_b1)
    plate_path_b2 = mirror.get(remote_plate_path_b2)
    plate_path_r1 = mirror.get(remote_plate_path_r1)
    plate_path_r2 = mirror.get(remote_plate_path_r2)

    hdulist_b1 = fitsio.FITS(plate_path_b1, mode=fitsio.READONLY)
    hdulist_b2 = fitsio.FITS(plate_path_b2, mode=fitsio.READONLY)
    hdulist_r1 = fitsio.FITS(plate_path_r1, mode=fitsio.READONLY)
    hdulist_r2 = fitsio.FITS(plate_path_r2, mode=fitsio.READONLY)
    
    TraceSet_b1 = bossdata.plate.TraceSet(hdulist_b1[3])
    TraceSet_b2 = bossdata.plate.TraceSet(hdulist_b2[3])
    TraceSet_r1 = bossdata.plate.TraceSet(hdulist_r1[3])
    TraceSet_r2 = bossdata.plate.TraceSet(hdulist_r2[3])

    return TraceSet_b1, TraceSet_b2, TraceSet_r1, TraceSet_r2

def combine(plate, mjd, fwhm=True):
    
    '''Combine data.
    
    
    Parameters:
    -----------
    plate: list/array
    mjd: list/array
    fwhm: rms or fwhm
    
    '''
    
    fwhm = np.sqrt(8*np.log(2))
    coef = 2*np.sqrt(np.pi)
    
    
        
    # Profile width data
    b1neff = np.empty((len(plate), 500, 4112), dtype=float)
    b2neff = np.empty((len(plate), 500, 4112), dtype=float)
    r1neff = np.empty((len(plate), 500, 4128), dtype=float)
    r2neff = np.empty((len(plate), 500, 4128), dtype=float)        

    # Wavelength data
    b1wlen = np.empty((len(plate), 500, 4128), dtype=float)
    b2wlen = np.empty((len(plate), 500, 4128), dtype=float)
    r1wlen = np.empty((len(plate), 500, 4128), dtype=float)
    r2wlen = np.empty((len(plate), 500, 4128), dtype=float)

    # Wavelength dispersion data
    b1res = np.empty((len(plate), 500, 4128), dtype=float)
    b2res = np.empty((len(plate), 500, 4128), dtype=float)
    r1res = np.empty((len(plate), 500, 4128), dtype=float)
    r2res = np.empty((len(plate), 500, 4128), dtype=float)

    # Row width data
    b1width = np.empty((len(plate), 500, 4128), dtype=float)
    b2width = np.empty((len(plate), 500, 4128), dtype=float)
    r1width = np.empty((len(plate), 500, 4128), dtype=float)
    r2width = np.empty((len(plate), 500, 4128), dtype=float)
        
    
    
    for i in range(len(plate)):
        
        b1TraceSet, b2TraceSet, r1TraceSet, r2TraceSet = get_neff(plate=plate[i], mjd=mjd[i])

        b1neff[i] = b1TraceSet.get_y()
        b2neff[i] = b2TraceSet.get_y()
        r1neff[i] = r1TraceSet.get_y()
        r2neff[i] = r2TraceSet.get_y()
            
        b1, b2, r1, r2 = get_cframe_data(plate=plate[i], mjd=mjd[i])
        
        b1wlen[i] = 10**b1[3].data
        b2wlen[i] = 10**b2[3].data
        r1wlen[i] = 10**r1[3].data
        r2wlen[i] = 10**r2[3].data
        
        b1width[i] = np.gradient(b1wlen[i], axis=1)
        b2width[i] = np.gradient(b2wlen[i], axis=1)
        r1width[i] = np.gradient(r1wlen[i], axis=1)
        r2width[i] = np.gradient(r2wlen[i], axis=1)

        b1res[i] = b1[4].data*0.0001*np.log(10)*b1wlen[i]
        b2res[i] = b2[4].data*0.0001*np.log(10)*b2wlen[i]
        r1res[i] = r1[4].data*0.0001*np.log(10)*r1wlen[i]
        r2res[i] = r2[4].data*0.0001*np.log(10)*r2wlen[i]
                
        if fwhm:
        
            b1res[i] = b1res[i]*fwhm
            b2res[i] = b2res[i]*fwhm
            r1res[i] = r1res[i]*fwhm
            r2res[i] = r2res[i]*fwhm
                
                
    return(b1wlen, b2wlen, r1wlen, r2wlen, b1res, b2res, r1res, r2res, 
           b1width, b2width, r1width, r2width, coef*b1neff, coef*b2neff, coef*r1neff, coef*r2neff)


def define_canon_grid(bwlen, rwlen, n_out, lo_trim=None, hi_trim=None):
    '''Define a canonical wavelength grid based on minimum and maximum wavlength range of all fibers on a plate.'''
    
    b_canon = np.linspace(np.max(bwlen[:,:,0]), np.min(bwlen[:,:,-1]), n_out)
    r_canon = np.linspace(np.max(rwlen[:,:,0]), np.min(rwlen[:,:,-1]), n_out)
    
    if lo_trim:
        b_canon = b_canon[lo_trim:]
        r_canon = r_canon[lo_trim:]
        
    if hi_trim:
        b_canon = b_canon[:hi_trim]
        r_canon = r_canon[:hi_trim]
    
    return(b_canon, r_canon)

def summarize(b_grid, r_grid, b_wlen, r_wlen, b_data, r_data, summary=np.median):
    
    assert b_wlen.shape == b_data.shape
    assert r_wlen.shape == r_data.shape
    
    n_out = len(b_canon) # red and blue have different lengths
    num_plates = b_data.shape[0]
        
    b_interp = np.empty((num_plates, 1000, n_out), dtype=float)
    r_interp = np.empty((num_plates, 1000, n_out), dtype=float)
    
    for plate in range(num_plates):
        for fiber in range(1000):
            b_interp[plate, fiber] = np.interp(x=b_grid, xp=b_wlen[plate][fiber], fp=b_data[plate][fiber])
            r_interp[plate, fiber] = np.interp(x=r_grid, xp=r_wlen[plate][fiber], fp=r_data[plate][fiber])
    
    return(summary(b_interp, axis=(0,1)), summary(r_interp, axis=(0,1)))
```

```python
plate = [7027, 6963, 7301, 6759, 6002, 6178, 6626, 6882, 7389, 7453, 7517, 6472, 6660, 6877, 6970, 6122, 7456, 7377, 
        7454, 7564]
        
mjd = [56448, 56724, 56746, 56416, 56104, 56213, 56330, 56541, 56769, 56749, 56772, 56362, 56370, 56544, 56444, 
       56246, 56727, 56741, 56751, 56804]

b1wlen, b2wlen, r1wlen, r2wlen, b1res, b2res, r1res, r2res, b1width, b2width, r1width, r2width, b1neff, b2neff, r1neff, r2neff = combine(plate, mjd, fwhm=True)

# Combine all fibers for each camera
bwlen = np.concatenate((b1wlen, b2wlen), axis=1)
rwlen = np.concatenate((r1wlen, r2wlen), axis=1)
bres = np.concatenate((b1res, b2res), axis=1)
rres = np.concatenate((r1res, r2res), axis=1)
bwidth = np.concatenate((b1width, b2width), axis=1)
rwidth = np.concatenate((r1width, r2width), axis=1)
bneff = np.concatenate((b1neff, b2neff), axis=1)
rneff = np.concatenate((r1neff, r2neff), axis=1)

# Create canonical wavelength grid
n_out = bneff.shape[2]
b_canon, r_canon = define_canon_grid(bwlen, rwlen, n_out, lo_trim=60, hi_trim=-80)
```

```python
bres_int, rres_int = summarize(b_canon, r_canon, bwlen, rwlen, bres, rres)
bwid_int, rwid_int = summarize(b_canon, r_canon, bwlen, rwlen, bwidth, rwidth)
bneff_int, rneff_int = summarize(b_canon, r_canon, bwlen[:,:,:n_out], rwlen, bneff, rneff)
```

### Adjust additional data files to be within interpolation range

```python
def adjust_interp(data, wlen_min, wlen_max, col_name):
    beg = np.arange(wlen_min, 3500, 0.1)
    end = np.arange(10000, wlen_max, 0.1)
    col_beg = np.full(len(beg), data[col_name][0])
    col_end = np.full(len(end), data[col_name][-1])
    pad_beg = Table([beg, col_beg], names=('WAVELENGTH', col_name))
    pad_end = Table([end, col_end], names=('WAVELENGTH', col_name))
    new_data = vstack([pad_beg, data])
    new_data = vstack([new_data, pad_end])
    return(new_data)
```

```python
def get_frame_flat_files(plate, mjd, fiber, band, sequence_number=0):
    plan_path = finder.get_plate_plan_path(plate=plate, mjd=mjd, combined=True)
    local_plan_path = mirror.get(plan_path)
    Plan = bossdata.plate.Plan(local_plan_path)


    # Get frame file
    expname_frame = Plan.get_exposure_name(sequence_number=sequence_number, band=band, fiber=fiber, 
                                           ftype='spFrame')
    local_frame = mirror.get(os.path.join(plan_path.strip(os.path.basename(plan_path)), 
                                         expname_frame))
    
    # Get cframe file
    expname_cframe = Plan.get_exposure_name(sequence_number=sequence_number, band=band, fiber=fiber, 
                                           ftype='spCFrame')
    local_cframe = mirror.get(os.path.join(plan_path.strip(os.path.basename(plan_path)), 
                                         expname_cframe))
    
    # Get flat file
    spec_index = Plan.get_spectrograph_index(fiber=fiber)
    FrameFile = bossdata.plate.FrameFile(path=local_frame, index=spec_index, calibrated=False)
    
    plate_spec_path = finder.get_plate_spec_path(plate=plate, mjd=mjd)
    local_plate_spec_path = mirror.get(plate_spec_path)
    PlateFile = bossdata.plate.PlateFile(local_plate_spec_path)
    exp_name_plate = PlateFile.exposures.get_exposure_name(exposure_index=sequence_number, 
                                                      camera=FrameFile.camera, ftype='spFlat')
    remote_plate_path = finder.get_plate_path(plate=plate, filename=exp_name_plate)
    plate_path = mirror.get(remote_plate_path)
    
    return(fits.open(local_frame), fits.open(local_cframe), fits.open(plate_path))
```

```python
def get_response(frame, flat, cframe, fiber, nom_wlen, extinct=new_ext):
    
    ff = np.mod(fiber, 500)
    
    eflux = frame[0].data[ff] # flux in flat-fielded electrons
    fiberflat = flat[0].data[ff]
    superflat = frame[8].data[ff]
    flux = cframe[0].data[ff]
    electrons = eflux * superflat * fiberflat
    exp_time = cframe[0].header['EXPTIME'] #s
    wlen = 10**cframe[3].data[ff] #check to see if matches frame[3].data
        
    flux = flux[:len(electrons)]
    wlen = wlen[:len(electrons)]
                
    flux_interp = np.interp(x=nom_wlen, xp=wlen, fp=flux)
    elec_interp = np.interp(x=nom_wlen, xp=wlen, fp=electrons)
    extinct_interp = np.interp(x=nom_wlen, xp=extinct['WAVELENGTH'], 
                               fp=extinct['EXTINCTION'])
    
    
    airmass = cframe[0].header['AIRMASS']
    exp_time = cframe[0].header['EXPTIME']
    
    midpt = np.append((nom_wlen[1:]+nom_wlen[:-1])/2, nom_wlen[-1])
    
    r = 250./2 # cm
    mirror_area = np.pi*r**2 # cm^2 
    h = scipy.constants.Planck
    c = scipy.constants.c
    
    dlam = np.diff(nom_wlen)
    dlam = np.append(dlam, dlam[-1])
    
    
    num = flux_interp * mirror_area * exp_time * dlam 
    denom = h * c / midpt
    photons = num / denom * (10**(-34)) # convert from Joules to ergs

    exp = extinct_interp*airmass
    extinction = 10 ** (-exp/2.5)

    photons_ext = photons * extinction
    
    return(elec_interp/photons_ext)
```

```python
def plot_throughput_stages(plate, mjd, fiber, band, wlen_min, wlen_max, extinct=new_ext, plot=True, save=None):
    
    frame, cframe, flat = get_frame_flat_files(plate=plate, mjd=mjd, fiber=fiber, band=band)
    
    # data indices start at 0, and go from 0-499; frames/flats fibers take values from 1-1000
    
    if fiber < 501:
        ff = np.mod(fiber, 501) - 1
    else:
        ff = np.mod(fiber, 501)
    
    
    eflux = frame[0].data[ff] # flux in flat-fielded electrons
    fiberflat = flat[0].data[ff]
    superflat = frame[8].data[ff]
    flux = cframe[0].data[ff]
    electrons = eflux * superflat * fiberflat
    exp_time = cframe[0].header['EXPTIME'] #s
    wlen = 10**cframe[3].data[ff] #check to see if matches frame[3].data
    
    flux = flux[:len(electrons)]
    wlen = wlen[:len(electrons)]
    
    if band == 'blue':
        nom_wlen = b_canon
    else:
        nom_wlen = r_canon
        
    flux_interp = np.interp(x=nom_wlen, xp=wlen, fp=flux)
    elec_interp = np.interp(x=nom_wlen, xp=wlen, fp=electrons)
    extinct_interp = np.interp(x=nom_wlen, xp=extinct['WAVELENGTH'], 
                               fp=extinct['EXTINCTION'])
    
    superflat_interp = np.interp(x=nom_wlen, xp=wlen, 
                               fp=superflat)
    fiberflat_interp = np.interp(x=nom_wlen, xp=wlen, 
                               fp=fiberflat)
    
    
    airmass = cframe[0].header['AIRMASS']
    exp_time = cframe[0].header['EXPTIME']
    
    midpt = np.append((nom_wlen[1:]+nom_wlen[:-1])/2, nom_wlen[-1])
    
    r = 250./2 # cm
    mirror_area = np.pi*r**2 # cm^2 
    h = scipy.constants.Planck
    c = scipy.constants.c
    
    dlam = np.diff(nom_wlen)
    dlam = np.append(dlam, dlam[-1])
    
    
    num = flux_interp * mirror_area * exp_time * dlam 
    denom = h * c / midpt
    photons = num / denom * (10**(-34)) # convert from Joules to ergs

    exp = extinct_interp*airmass
    extinction = 10 ** (-exp/2.5)

    photons_ext = photons * extinction
    
    if plot:
    
        fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6), (ax7, ax8)) = plt.subplots(4,2, figsize=(11,11))


        ax1.scatter(nom_wlen, flux_interp, alpha=0.3, s=1)
        ax1.set_title('Calibrated flux')
        ax1.set_xlabel('Wavelength [Angstrom]')
        ax1.set_ylabel(r'Flux $[10^{-17}$ erg/s/cm$^2$/A]')
        ax1.set_xlim(wlen_min, wlen_max)
        flux_min, flux_max = np.percentile(flux_interp, (1, 99))
        ax1.set_ylim(0, flux_max)

        ax2.scatter(nom_wlen, photons, alpha=0.3, s=1)
        ax2.set_title('Photons entering atmosphere')
        ax2.set_xlabel('Wavelength [Angstrom]')
        ax2.set_ylabel('# Photons')
        ax2.set_xlim(wlen_min, wlen_max)
        photon_min, photon_max = np.percentile(photons, (1,99))
        ax2.set_ylim(0, photon_max)

        ax7.scatter(nom_wlen, elec_interp, alpha=0.3, s=1)
        ax7.set_title('Detected electrons')
        ax7.set_xlabel('Wavelength [Angstrom]')
        ax7.set_ylabel('# Electrons')
        ax7.set_xlim(wlen_min, wlen_max)
        elec_min, elec_max = np.percentile(elec_interp, (1,99))
        ax7.set_ylim(0, elec_max+50)

        ax3.scatter(nom_wlen, extinction, alpha=0.3, s=1)
        ax3.set_title('Extinction')
        ax3.set_xlabel('Wavelength [Angstrom]')
        ax3.set_ylabel('Extinction')
        ax3.set_xlim(wlen_min, wlen_max)
        ex_min, ex_max = np.percentile(extinct['EXTINCTION'], (1,99))
        ax3.set_ylim(0, 1.1)

        ax5.scatter(nom_wlen, superflat_interp, alpha=0.3, s=1)
        ax5.set_title('Superflat')
        ax5.set_xlabel('Wavelength [Angstrom]')
        ax5.set_ylabel('# Electrons')
        ax5.set_xlim(wlen_min, wlen_max)
        sup_min, sup_max = np.percentile(superflat_interp, (1,99))
        ax5.set_ylim(0, 1.8)

        ax6.scatter(nom_wlen, fiberflat_interp, alpha=0.3, s=1)
        ax6.set_title('Fiberflat')
        ax6.set_xlabel('Wavelength [Angstrom]')
        ax6.set_ylabel('# Electrons')
        ax6.set_xlim(wlen_min, wlen_max)
        ax6.set_ylim(0, 1.1)

        ax4.scatter(nom_wlen, photons_ext, alpha=0.3, s=1)
        ax4.set_title('Electrons entering telescope')
        ax4.set_xlabel('Wavelength [Angstrom]')
        ax4.set_ylabel('# Electrons')
        ax4.set_xlim(wlen_min, wlen_max)
        photons_ext_min, photons_ext_max = np.percentile(photons_ext, (1,99))
        ax4.set_ylim(0, photons_ext_max)

        ax8.scatter(nom_wlen, elec_interp/photons_ext, alpha=0.3, s=1)
        ax8.set_title('Fiberloss x Throughput')
        ax8.set_xlabel('Wavelength [Angstrom]')
        ax8.set_ylabel('# Electrons / # Photons')
        ax8.set_xlim(wlen_min, wlen_max)
        ax8.set_ylim(0,0.25)


        plt.suptitle('Standard Star \n PLATE={}, MJD={}, FIBER={}, CAMERA={}'.format(plate, mjd, fiber, band), 
                     fontsize=12)
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])


        if save:
                plt.savefig(save)

        plt.show()
    
    return(elec_interp/photons_ext, flux_interp)
```

```python
def all_plate_thru(standards):
        
    plate = standards['PLATE'][0]
    mjd = standards['MJD'][0]
    star = standards['CLASS'][0]
    
    assert np.all(standards['PLATE'] == plate)
    assert np.all(standards['MJD'] == mjd)
    assert np.all(standards['CLASS'] == star)
        
    less_than = standards['FIBER'] < 500
    greater_than = ~less_than
    
    cam1 = standards[less_than]
    cam2 = standards[greater_than]

    thru_b1 = np.empty((len(cam1), len(b_canon)), dtype=float)
    thru_r1 = np.empty((len(cam1), len(r_canon)), dtype=float)
    thru_b2 = np.empty((len(cam2), len(b_canon)), dtype=float)
    thru_r2 = np.empty((len(cam2), len(r_canon)), dtype=float)
    
    for i, ff in enumerate(cam1['FIBER']):
        frame, cframe, flat = get_frame_flat_files(plate=plate, mjd=mjd, fiber=ff, band='blue')
        thru_b1[i] = get_response(frame, flat, cframe, ff, nom_wlen=b_canon, extinct=new_ext)
        frame, cframe, flat = get_frame_flat_files(plate=plate, mjd=mjd, fiber=ff, band='red')
        thru_r1[i] = get_response(frame, flat, cframe, ff, nom_wlen=r_canon, extinct=new_ext)
    
    for j, ff in enumerate(cam2['FIBER']):
        frame, cframe, flat = get_frame_flat_files(plate=plate, mjd=mjd, fiber=ff, band='blue')
        thru_b2[j] = get_response(frame, flat, cframe, ff, nom_wlen=b_canon, extinct=new_ext)
        frame, cframe, flat = get_frame_flat_files(plate=plate, mjd=mjd, fiber=ff, band='red')
        thru_r2[j] = get_response(frame, flat, cframe, ff, nom_wlen=r_canon, extinct=new_ext)

    all_b = np.concatenate((thru_b1, thru_b2), axis=0)
    all_r = np.concatenate((thru_r1, thru_r2), axis=0)
    
    return(all_b, all_r)
```

```python
def get_std_params(standards):

    plate = standards['PLATE'][0]
    mjd = standards['MJD'][0]
    fiber_num = standards['FIBER']
    
    frame1, cframe1, flat1 = get_frame_flat_files(plate=plate, mjd=mjd, fiber=1, band='blue')
    frame2, cframe2, flat2 = get_frame_flat_files(plate=plate, mjd=mjd, fiber=501, band='blue')
    

    focal_pos = np.empty((len(standards), 2), dtype=float)
    cart_num = np.empty((len(standards), 1), dtype=float)
    am =  np.empty((len(standards), 1), dtype=float)
    et = np.empty((len(standards), 1), dtype=float)
    
    cam1 = standards[standards['FIBER'] < 501]
    cam2 = standards[standards['FIBER'] > 500]
    
    cam1_fibers = np.mod(cam1['FIBER'], 501) - 1
    cam2_fibers = np.mod(cam2['FIBER'], 501)
        
    for i, ff in enumerate(cam1_fibers):
            
        focal_pos[i][0] = cframe1[5].data['XFOCAL'][ff]
        focal_pos[i][1] = cframe1[5].data['YFOCAL'][ff]
        cart_num[i] = cframe1[0].header['CARTID']
        am[i] = cframe1[0].header['AIRMASS']
        et[i] = cframe1[0].header['EXPTIME']
        
    split = len(cam1_fibers)
    
    for j, ff in enumerate(cam2_fibers):
        focal_pos[split + j][0] = cframe2[5].data['XFOCAL'][ff]
        focal_pos[split + j][1] = cframe2[5].data['YFOCAL'][ff]
        cart_num[split + j] = cframe2[0].header['CARTID']
        am[split + j] = cframe2[0].header['AIRMASS']
        et[split + j] = cframe2[0].header['EXPTIME']
    
    
    return(focal_pos, fiber_num, cart_num, am, et)
        

def get_features(thru, total_flux, focal_pos):
    num = thru.shape[0]
    mean_thru = np.empty(num, dtype=float)
    mean_flux = np.empty(num, dtype=float)
    for i in range(num):
        mean_thru[i] = np.mean(thru[i])
        mean_flux[i] = np.mean(total_flux[i])
    x = focal_pos[:,0]
    y = focal_pos[:,1]
    r = np.sqrt(x ** 2 + y ** 2)
    return(mean_thru, mean_flux, r)
```
