### Put into functions

```python
def get_frame_file(plan_path, sequence_number, band, fiber):
    local_plan_path = mirror.get(plan_path)
    Plan = bossdata.plate.Plan(local_plan_path)
    exp_name_plan = Plan.get_exposure_name(sequence_number=sequence_number, band=band, fiber=fiber, 
                                           ftype='spFrame')
    local_fits = mirror.get(os.path.join(plan_path.strip(os.path.basename(plan_path)), 
                                         exp_name_plan))
    spec_index = Plan.get_spectrograph_index(fiber=fiber)
    FrameFile = bossdata.plate.FrameFile(path=local_fits, index=spec_index, calibrated=False)
    return fits.open(local_fits), FrameFile 

def get_flat_file(plate, mjd, sequence_number, camera):
    remote_plate_spec_path = finder.get_plate_spec_path(plate=plate, mjd=mjd)
    plate_spec_path = mirror.get(remote_plate_spec_path)
    PlateFile = bossdata.plate.PlateFile(plate_spec_path)
    exp_name_plate = PlateFile.exposures.get_exposure_name(exposure_index=sequence_number, 
                                                      camera=FrameFile.camera, ftype='spFlat')
    remote_plate_path = finder.get_plate_path(plate=plate, filename=exp_name_plate)
    plate_path = mirror.get(remote_plate_path)
    return fits.open(plate_path)

def get_cframe_file(plan_path, sequence_number, band, fiber):
    local_plan_path = mirror.get(plan_path)
    Plan = bossdata.plate.Plan(local_plan_path)
    exp_name_plan = Plan.get_exposure_name(sequence_number=sequence_number, band=band, fiber=fiber, 
                                           ftype='spCFrame')
    local_fits = mirror.get(os.path.join(plan_path.strip(os.path.basename(plan_path)), 
                                         exp_name_plan))
    return fits.open(local_fits)
```

```python
def get_flats(plate, mjd, fiber, sequence_number, band):
    plan_path = finder.get_plate_plan_path(plate=plate, mjd=mjd, combined=True)
    local_plan_path = mirror.get(plan_path)
    frame, FrameFile = get_frame_file(plan_path=plan_path, sequence_number=sequence_number, 
                                  band=band, fiber=fiber)
    flat = get_flat_file(plate=plate, mjd=mjd, sequence_number=sequence_number, 
                         camera=FrameFile.camera)
    cframe = get_cframe_file(plan_path=plan_path, sequence_number=sequence_number, 
                             band=band, fiber=fiber)
    return frame[8].data, flat[0].data #(super,fiber)
```

## Calculate fiberloss x throughput as a function of wavelength

1) Make a plot of F(lam) x T(lam) for all fibers on a plate
    - Sanity check: should be between 0 and 1
    - This should be varying, and I would imagine fiberloss would be less for stars/quasars
2) Make plot of fiberloss as a function of wavelength, where fiberloss is generated by specsim

http://specsim.readthedocs.io/en/stable/guide.html


According to Stephen's note on `From flux to photons`

https://trac.sdss3.org/wiki/BOSS/pipeline/FluxToPhotons 


`flux = electrons x calib`

- where electrons is in # of electrons, and calib has units of flux (ergs/s/cm2/A)
    
    
The fiberloss x throughput is equal to the (#electrons detected/#photons) that enter the telescope
 - note: throughput includes reflectivity of the primary (and secondary?) mirror
 
The number of photons that enter the telescope can be extracted from the electron flux.

##### Make a function that takes (plate,mjd,fiber,bin_size) and returns fiberloss x throughput

```python
plate = 7027 
mjd = 56448
fiber = 10
sequence_number = 0
band = 'red'
```

```python
plan_path = finder.get_plate_plan_path(plate=plate, mjd=mjd, combined=True)
local_plan_path = mirror.get(plan_path)

frame, FrameFile = get_frame_file(plan_path=plan_path, sequence_number=sequence_number, 
                                  band=band, fiber=fiber)
flat = get_flat_file(plate=plate, mjd=mjd, sequence_number=sequence_number, 
                     camera=FrameFile.camera)
cframe = get_cframe_file(plan_path=plan_path, sequence_number=sequence_number, 
                         band=band, fiber=fiber)

eflux = frame[0].data # flux in flat-fielded electrons (10^{-17} erg/s/cm^2/A)
superflat = frame[8].data
fiberflat = flat[0].data
electrons = eflux * superflat * fiberflat

flux = cframe[0].data
calib = flux/electrons
```
